```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("Diwali Sales Data.csv", encoding= 'unicode_escape')
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
data.info()
```

```python
data.shape
```

```python
pd.isnull(data).sum()
```

```python
data.dropna(inplace=True)
```

```python
pd.isnull(data).sum()
```

```python
data["Amount"] = data["Amount"].astype(int)
```

```python
data["Amount"].dtype
```

```python
data.columns
```

```python
data.rename(columns={"Cust_name" : "Customer Name"})
```

```python
data[["Age", "Amount"]].describe()
```

```python
ax = sns.countplot(x="Gender", data=data, palette="Set2")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_gen = data.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

sns.barplot(x = "Gender", y = "Amount", data = sales_gen, palette="Set2")
```

```python
sales_age = data.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

sns.barplot(x = "Age Group", y = "Amount", data = sales_age, palette="Set2")
```

```python
ax = sns.countplot(data = data, x = "Age Group", hue = "Gender")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
data.columns
```

```python
ax = sns.countplot(x="Marital_Status", data=data, palette="Set2")
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = data.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
plt.figure(figsize=(15,7))
sns.barplot(x = "State", y = "Amount", data = sales_state, palette="Set2")
```

```python
sales_state = data.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False).head(10)
plt.figure(figsize=(15,7))
sns.barplot(x = "State", y = "Orders", data = sales_state, palette="Set2")
```

```python
sales_state = data.groupby(["Marital_Status", "Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
plt.figure(figsize=(3, 5))
sns.barplot(x = "Marital_Status", y = "Amount", hue = "Gender", data = sales_state, palette="Set2")
```

```python
data.columns
```

```python
sales_occ = data.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
plt.figure(figsize=(20,8))
sns.barplot(x = "Occupation", y = "Amount", data = sales_occ, palette="Set2")
```

```python
ax = sns.countplot(x="Occupation", data=data, palette="Set2")
sns.set(rc={"figure.figsize":(25, 10)})

for bars in ax.containers:
    ax.bar_label(bars)
```
